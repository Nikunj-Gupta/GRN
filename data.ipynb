```python
import GEOparse 
import pandas as pd 
```

```python
# gse = GEOparse.get_GEO(geo="GSE1563", destdir="./")
gse = GEOparse.get_GEO(filepath="./datasets/GSE32591_family.soft.gz")
# gse = GEOparse.get_GEO(filepath="./datasets/soft_files/GSE45291_family.soft.gz")
```

```python
print("GSM example:")
for gsm_name, gsm in gse.gsms.items():
    print("Name: ", gsm_name)
    print("Metadata:",)
    for key, value in gsm.metadata.items():
        print(" - %s : %s" % (key, ", ".join(value)))
    print ("Table data:",)
    print (gsm.table.head())
    print (len(gsm.table['ID_REF'].unique()))
    print(gsm.table.columns)
    break 
```

```python
print("GPL example:")
for gpl_name, gpl in gse.gpls.items():
    print("Name: ", gpl_name)
    print("Metadata:",)
    for key, value in gpl.metadata.items():
        print(" - %s : %s" % (key, ", ".join(value)))
    print("Table data:",)
    print(gpl.table) 
    print(gpl.table.columns)
    break
```
